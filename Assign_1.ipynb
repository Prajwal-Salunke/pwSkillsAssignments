Q5.if A is purely divisible by B

```python
A = int(input("Enter the number A:- "))
B = int(input("Enter the number B:- "))
if A % B == 0:
    print(f"{A} is purely divisible by {B} ")
    while A % B == 0:
        print(f"{A} is {int(A / B)} times divisible by {B}")
        break
else:
    print(f"{A} is NOT purely divisible by {B} ")
```

Q6. Divisible by 3

```python
list2 = []
for i in range(25, 51):
    list2.append(i)
```

```python
for j in list2:
    if j % 3 == 0:
        print(f"{j} is divisible by 3 ")
    else:
        print(f"{j} is NOT divisible by 3 ")
```

Q7. MUTABLE & IMMUTABLE Data types


<br>
Mutable Data types :- LIST , SET , Dictionary<br>
    <br>
    In these data types ,we can change(add/delete) the value.<br>
    Example :- <br>


```python
list3 = ["apple", "rose", 2, 5]
print("Original List :- ", list3)
list3[1] = "banana"  # "rose" is changed with "banana"
print("Mutable List:- ", list3)
```


<br>
Immutable data types :- String , Tuple<br>
    The value associated with these data types can not be changed  once set. We can not add/delete or update the values<br>
    Example:-<br>
string1="Python"<br>
string1[2]= "l"    # NOT POSSIBLE ...Error will occurred at this line<br>
tuple1=(5,8,66,0)<br>
